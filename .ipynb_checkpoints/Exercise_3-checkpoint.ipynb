## Movie Dataset

```python
movie_df = pd.read_csv('movies.csv', sep=',')
```

## Range of movie gross

```python
gross = movie_df.iloc[:,1].tolist()
max_gross = 0
min_gross = 99999999
zero_count = 0
for movie in gross:
    try:
        val = float(movie)
        if val > max_gross:
            max_gross = val
        elif val < min_gross and val > 0:
            min_gross = val
        elif val == 0:
            zero_count += 1

    except ValueError:
        pass
print(f'The lowest gross (except movies witt $0 gross) for a movie in the USA was ${min_gross} and the max gross was ${max_gross}.')
print(f'The amout of movies with $0 gross was {zero_count}.')
```

## Most succesful type of movie based on world wide gross

```python
movies = {}
for i in range(movie_df.shape[0]):
    rating = movie_df.iloc[i,6]
    genre = movie_df.iloc[i,10]
    gross = movie_df.iloc[i,2]
    if gross != 'Unknown':
        gross = float(gross)
        if genre != 'nan':
            if genre in movies:
                if movies[genre] < gross:
                    movies[gross] = gross
            else:
                movies[genre] = gross
        if rating != 'nan' and rating != 'Not Rated':
            if rating in movies:
                if movies[rating] < gross:
                    movies[rating] = gross
            else:
                movies[rating] = gross
best_type = None
highest_gross = 0
for label in movies:
    if movies[label] > highest_gross:
        highest_gross = movies[label]
        best_type = label
print(f'Based on world wide gross {best_type} is the most succesful type of movie with a gross of ${highest_gross}')
```

## GDP Per Capita Dataset

```python
gdp = pd.read_csv('gapdata.csv', sep=';')
gdp = gdp.iloc[:,0:3]
```

```python
gdp.iloc[:,2] = gdp.iloc[:,2].str.replace(' ', '')
gdp.iloc[:,2] = pd.to_numeric(gdp.iloc[:,2])
```

```python
updated_gdp = gdp
updated_gdp["Continent"] = ""
```

```python
for i in range(updated_gdp.shape[0]):
    try:
        name = updated_gdp.iloc[i,0]
        country_code = pc.country_name_to_country_alpha2(name, cn_name_format="default")
        continent_name = pc.country_alpha2_to_continent_code(country_code)
        updated_gdp.at[i, 'Continent'] = continent_name
    except KeyError:
        #updated_gdp=updated_gdp.drop(updated_gdp.index[i])
        pass
```

```python
updated_gdp = updated_gdp[updated_gdp.iloc[:,3] != '']
del updated_gdp["Area"]
```

```python
gdp_asia = updated_gdp[updated_gdp.iloc[:,2] == 'AS']
gdp_europe = updated_gdp[updated_gdp.iloc[:,2] == 'EU']
gdp_africa = updated_gdp[updated_gdp.iloc[:,2] == 'AF']
```

```python
years = sqr = list(range(1270, 2022, 20))
plt.figure(figsize=(15,8))
plt.plot( 
    gdp_asia.groupby('Year').mean().index,
    gdp_asia.groupby('Year').mean()
)

plt.plot( 
    gdp_europe.groupby('Year').mean().index,
    gdp_europe.groupby('Year').mean()
)

plt.plot( 
    gdp_africa.groupby('Year').mean().index,
    gdp_africa.groupby('Year').mean()
)

plt.xticks(years)
plt.xticks(rotation = 45)
plt.ylabel('GDP per capita')
plt.xlabel('Year')
plt.legend(['Africa', 'Europe', 'Africa'])
```
